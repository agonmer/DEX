```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the files containing data from the two samples for the U tests.

file_a = 'sample_a_data_FC.csv'
dfa = pd.read_csv(file_a,sep='\t')
file_b = 'sample_b_data_FC.csv'
dfb = pd.read_csv(file_b,sep='\t')
```

```python
dfa.columns
```

```python
valid_fields = dfa.columns

for kk in range(len(valid_fields)):
    a = dfa[valid_fields[kk]]
    b = dfb[valid_fields[kk]]
    if valid_fields[kk]=='FC_RR3':
        a,b = a[a>0].copy(),b[a>0].copy()
    elif valid_fields[kk]=='FC_RR3-ent':
        a,b = a[a>-1000].copy(),b[a>-1000].copy()
    elif valid_fields[kk]=='perc FC_RR3-ent':
        a,b = a[a>-1000].copy(),b[a>-1000].copy()

    res = stats.mannwhitneyu(a,b)
    
    print(valid_fields[kk])
    print(res)
    print('------------------------------------------------------------------------------')

    fig = plt.figure(figsize=(8,6))
    plt.hist(a,alpha=0.5,label='D')
    plt.hist(b,alpha=0.5,label='K')
    plt.legend(fontsize=14)
    plt.tick_params(axis='both',labelsize=14)
    plt.xlabel(valid_fields[kk],fontsize=14)
    plt.title('Mann Whitney U test p value = {pval:.3f}'.format(pval = res.pvalue),fontsize=16)
    if valid_fields[kk]=='TOT PAED_RR0':
        figure_name = 'histogram_tot_paed_rr0.png'
        plt.savefig(figure_name,bbox_inches='tight',r=300)
```

```python
# Build the time series and plot them.

mean_a = [] # Mean values.
mean_b = []
stem_a = [] # Standard error of the mean values.
stem_b = []
instants = ['ent','T0','T1','T2','T3','T4','fin','eot']

for ll in range(len(instants)):
    
    set_a = dfa['FC_'+instants[ll]]
    set_b = dfb['FC_'+instants[ll]]
    
    mean_a.append(np.mean(set_a))    
    mean_b.append(np.mean(set_b))
    stem_a.append(np.std(set_a)/np.sqrt(len(set_a)))    
    stem_b.append(np.std(set_b)/np.sqrt(len(set_b)))    
    
fig = plt.figure(figsize=(8,6))
plt.errorbar(instants,mean_a,stem_a,fmt='o-',label='D')
plt.errorbar(instants,mean_b,stem_b,fmt='v-',label='K')
plt.legend(fontsize=14,loc='upper left')
plt.ylabel('Heart rate [bpm]',fontsize=14)
plt.xlabel('Time',fontsize=14)
plt.tick_params(axis='both',labelsize=14)
plt.grid()

figure_name = 'FC_timeseries.png'
plt.savefig(figure_name,bbox_inches='tight',r=300)
```
