```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress
```

```python
# Read the file containing data for the relationship entropia-gas.
# Both samples can be considered.

sample = 'a'

df = pd.read_csv('sample_'+sample+'_entropia.csv',sep='\t')
```

```python
# Function to convert ',' in '.' and set the output to be a float
def Replace(str1):
    str1 = str1.replace(',', '.')
    return float(str1)
```

```python
valid_suff = ['T0', 'T1', 'T2', 'T3', 'T4', 'fin']

for kk in range(len(valid_suff)):
    suff = valid_suff[kk]
    xobj = df['EtSev_'+suff].copy()
    y = df['Entropia_'+suff].copy()
    
    x = np.zeros_like(xobj,dtype='float')
    for ll in range(len(xobj)):
        x[ll] = Replace(str(xobj[ll]))
    
    # Remove the outliers on Entropia above 150.
    x[y>150], y[y>150] = np.nan, np.nan
    x, y = x[~np.isnan(x)&~np.isnan(y)], y[~np.isnan(x)&~np.isnan(y)]
    
    fig = plt.figure()
    plt.scatter(x,y)
    plt.xlabel('EtSev',fontsize=14)
    plt.ylabel('Entropia',fontsize=14)
    
    slope, intercept, r, p, se = linregress(x, y)
    ics = np.array([np.min(x),np.max(x)])
    ipsilon = ics*slope+intercept
    plt.plot(ics,ipsilon,'r')
    plt.title(suff + ', slope = {s:.2f}'.format(s=slope) + ', corr.coeff. = {rho:.2f}'.format(rho=r),fontsize=16)
    
```
