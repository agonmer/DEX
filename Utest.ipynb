```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the files containing data from the two samples for the U tests.

file_a = 'sample_a_data.csv'
dfa = pd.read_csv(file_a,sep='\t')
file_b = 'sample_b_data.csv'
dfb = pd.read_csv(file_b,sep='\t')
```

```python
valid_fields = ['CV_RR0','AF_RR0','CON_RR0','3 ITEMS_RR0','IRR_RR0','INC_RR0','2 ITEMS_RR0','TOT PAED_RR0',
                'CV_RR1','AF_RR1','CON_RR1','IRR_RR1','INC_RR1','TOT PAED_RR1',
                'CV_RR2','AF_RR2','CON_RR2','IRR_RR2','INC_RR2','TOT PAED_RR2',
                'CV_RR3','AF_RR3','CON_RR3','IRR_RR3','INC_RR3','TOT PAED_RR3']

pval_u = []

for kk in range(len(valid_fields)):
    a = dfa[valid_fields[kk]]
    b = dfb[valid_fields[kk]]
    res = stats.mannwhitneyu(a,b)
    pval_u.append(res.pvalue)
    
    print(valid_fields[kk])
    print(res)
    print('------------------------------------------------------------------------------')
    
    fig = plt.figure(figsize=(8,6))
    plt.hist(a,alpha=0.5,label='D')
    plt.hist(b,alpha=0.5,label='K')
    plt.legend(fontsize=14)
    plt.tick_params(axis='both',labelsize=14)
    plt.xlabel(valid_fields[kk],fontsize=14)
    plt.title('Mann Whitney U test p value = {pval:.3f}'.format(pval = res.pvalue),fontsize=16)
    if valid_fields[kk]=='TOT PAED_RR0':
        figure_name = 'histogram_tot_paed_rr0.png'
        plt.savefig(figure_name,bbox_inches='tight',r=300)
    
print(pval_u)
```
